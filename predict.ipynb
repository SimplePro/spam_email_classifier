```python
import numpy as np
import pandas as pd
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./emails.csv")
df.head()
```

```python
x_df = df.drop(["Email No.", "Prediction"], axis=1)
x_df.head()
```

```python
y_df = df.Prediction
y_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df, test_size=0.2, random_state=156)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
lgbm = LGBMClassifier(n_estimators=100)
evals = [(X_test, y_test)]
lgbm.fit(X_train, y_train, early_stopping_rounds=100, eval_metric="logloss", eval_set=evals, verbose=True)
```

```python
pred = lgbm.predict(X_test)
print(pred)
```

```python
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

```python
import joblib
joblib.dump(lgbm, "spam_email.pkl")
```
